```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from a CSV file
df = pd.read_csv('/content/data.csv')
# Print a summary of the DataFrame
print(df.describe())

# Find the 25th, 50th, and 75th percentiles for the "Pulse" column
Pulse = df['Pulse']
percentile_25 = np.percentile(Pulse, 25)
percentile_50 = np.percentile(Pulse, 50)
percentile_75 = np.percentile(Pulse, 75)

# Print the percentiles
print(percentile_25)
print(percentile_50)
print(percentile_75)

# Calculate the standard deviation of the DataFrame
s_d = np.std(df)
print("Standart_Deviation:", s_d)

# Calculate the coefficient of variation
c_v = np.std(df) / np.mean(df)
print("Coefficient_Variation:", c_v)
# Calculate the variance of the DataFrame
var = np.var(df)
print("Variance:", var)

# Create a scatter plot of "Pulse" and "Maxpulse"
df.plot(x='Pulse', y='Maxpulse', kind='scatter')
plt.show()

```
